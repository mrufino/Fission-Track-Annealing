```python
import pandas as pd
import matplotlib.pyplot as plt

models = ["PA", "PC", "FA", "FC"]
rate_str = "1p0"  # ajuste para sua taxa, ex: "1p0" para 1.0 K/Ma
fontsize = 14     # controle o tamanho da fonte aqui

# Plot Delta Médio
plt.figure(figsize=(10, 6))
for model in models:
    df_delta = pd.read_csv(f"delta_profile_{model}_{rate_str}KMa.csv")
    plt.plot(df_delta['Tempo_Ma'], df_delta['Delta_Medio'], label=model)

plt.xlabel('Tempo (Ma)', fontsize=fontsize)
plt.ylabel('Delta Médio', fontsize=fontsize)
plt.title(f'Perfil Delta Médio - Taxa {rate_str.replace("p", ".")} K/Ma', fontsize=fontsize+2)
plt.grid(True)
plt.legend(title="Modelos", fontsize=fontsize, title_fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.tight_layout()
plt.savefig(f"delta_profile_all_{rate_str}KMa.png", dpi=300)
plt.show()

# Plot Gamma Médio
plt.figure(figsize=(10, 6))
for model in models:
    df_gamma = pd.read_csv(f"gamma_profile_{model}_{rate_str}KMa.csv")
    plt.plot(df_gamma['Tempo_Antes_Presente_Ma'], df_gamma['Gamma_Medio_Ma'], label=model)

plt.xlabel('Tempo Antes do Presente (Ma)', fontsize=fontsize)
plt.ylabel('Gamma Médio (Ma)', fontsize=fontsize)
plt.title(f'Perfil Gamma Médio - Taxa {rate_str.replace("p", ".")} K/Ma', fontsize=fontsize+2)
plt.grid(True)
plt.legend(title="Modelos", fontsize=fontsize, title_fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.tight_layout()
plt.savefig(f"gamma_profile_all_{rate_str}KMa.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

models = ["PA", "PC", "FA", "FC"]
rate_str = "1p0"
fontsize = 14

plt.figure(figsize=(10, 6))
for model in models:
    df_gamma = pd.read_csv(f"gamma_profile_{model}_{rate_str}KMa.csv")
    # Inverter o eixo x para mostrar de 250 a 0
    time_inv = 250 - df_gamma['Tempo_Antes_Presente_Ma']
    plt.plot(time_inv, df_gamma['Gamma_Medio_Ma'], label=model)

plt.xlabel('Tempo (Ma)', fontsize=fontsize)
plt.ylabel('Gamma Médio (Ma)', fontsize=fontsize)
plt.title(f'Perfil Gamma Médio - Taxa {rate_str.replace("p", ".")} K/Ma', fontsize=fontsize+2)
plt.grid(True)
plt.legend(title="Modelos", fontsize=fontsize, title_fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.gca().invert_xaxis()  # inverte o eixo x para que 250 fique à esquerda e 0 à direita
plt.tight_layout()
plt.savefig(f"gamma_profile_all_{rate_str}KMa_invertido.png", dpi=300)
plt.show()
```
